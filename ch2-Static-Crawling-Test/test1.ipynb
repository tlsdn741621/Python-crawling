```python
#bs4 (BeautifulSoup)
#   HTML XML 
#      .
#      
#          .

##  
#    (tag)    .
# BeautifulSoup      
#    (Parse Tree)  .
#              .
#
#  ,
#   '' ,
# BeautifulSoup  ()  ,
#    (, )    
#  '' .

# 
# BeautifulSoup  
#       .
#
# 1. BeautifulSoup  
#  HTML   soup   .
from bs4 import BeautifulSoup

#   HTML 
html_doc = "<html><head><title>The Dormouse's story</title></head><body>...</body></html>"

# BeautifulSoup   (HTML , ' ')
soup = BeautifulSoup(html_doc, 'html.parser')


# 2. (Tag) : find() find_all()
# soup.find(''):        .
# soup.find_all(''):      (List)  .
#    <p>   
p_tag = soup.find('p')
#  <li>  
li_tags = soup.find_all('li')

# 3  (Text) : .text
#   .text  
#        .
# <title>   
title_text = soup.find('title').text
print(title_text)  # : The Dormouse's story

# 4.  (Attribute)  
#  (:   href,   src)
#   []  .
# <a>  ,   href   
a_tag = soup.find('a')
link_url = a_tag['href']
print(link_url)

#  5 CSS  : select() select_one()
# CSS    select()   .
#
# soup.select_one(''): CSS      .
#
# soup.select(''): CSS      .
```

```python
import bs4
webPage = open(r'./Sample03.html','rt',encoding='utf-8').read()
bsObject2 = bs4.BeautifulSoup(webPage, "html.parser")
print(bsObject2)
```

```python
tag_id_myId1 = bsObject2.find('div', {'id': 'myId1'})
print(tag_id_myId1)
```

```python
tag_class_myClass1 = bsObject2.find('div', {'class': 'myClass1'})
print(tag_class_myClass1)
```

```python
tag_div_myClass1 = bsObject2.findAll('div', {'class': 'myClass1'})
print(tag_div_myClass1)
```

```python
ul_value = bsObject2.find('ul', {'class': 'myClass2'})
print(ul_value)
```

```python
li_list_myClass3 = bsObject2.findAll('li', {'class': 'myClass3'})
print(li_list_myClass3)
```

```python
a_list = bsObject2.findAll('a')
print(a_list)
```

```python
for aTag in a_list:
    print(aTag['href'])
```

```python
tag_li_all = bsObject2.findAll('li')
for tag_li in tag_li_all:
    print(tag_li.text)
```
